```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("processed_caltech.csv")
```

```python
df['connectionTime'] = pd.to_datetime(df['connectionTime'])
df['disconnectTime'] = pd.to_datetime(df['disconnectTime'])
df['doneChargingTime'] = pd.to_datetime(df['doneChargingTime'])
```

```python
df['hour'] = df['connectionTime'].dt.hour
df['day_of_week'] = df['connectionTime'].dt.dayofweek
df['month'] = df['connectionTime'].dt.month
```

```python
df[['hour','day_of_week','month']].head()
```

```python
hourly_demand = df.groupby('hour').size()

plt.figure(figsize=(12,5))
hourly_demand.plot(kind='bar')

plt.title("Charging Demand by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Sessions")

plt.show()
```

```python
weekday_demand = df.groupby('day_of_week').size()

plt.figure(figsize=(10,5))
weekday_demand.plot(kind='bar')

plt.title("Charging Demand by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Sessions")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.hist(
    df['kWhDelivered'],
    bins=40
)

plt.title("Energy Delivered Distribution")
plt.xlabel("kWh Delivered")
plt.ylabel("Frequency")

plt.show()
```

```python
top_stations = (
    df['stationID']
      .value_counts()
      .head(10)
)

plt.figure(figsize=(12,5))

top_stations.plot(kind='bar')

plt.title("Top 10 Busiest Stations")

plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("processed_caltech.csv")
```

```python
df['connectionTime'] = pd.to_datetime(df['connectionTime'])
```

```python
hourly_demand = df.groupby(
    df['connectionTime'].dt.hour
).size()

print(hourly_demand)
```

```python
import matplotlib.pyplot as plt

hourly_demand = df.groupby(
    df['connectionTime'].dt.hour
).size()

plt.figure(figsize=(10,5))

hourly_demand.plot(
    kind='line',
    marker='o'
)

plt.title("Hourly Charging Demand")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Sessions")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

hourly_demand.plot(
    kind='line',
    marker='o',
    linewidth=3
)

plt.axvspan(
    14,
    17,
    alpha=0.2,
    color='red',
    label='Peak Hours'
)

plt.title("Hourly Charging Demand Pattern")
plt.xlabel("Hour of Day")
plt.ylabel("Charging Sessions")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

hourly_demand = (
    df.groupby(df['connectionTime'].dt.hour)
    .size()
)

plt.figure(figsize=(12,6), facecolor='#0B3D3B')

ax = plt.gca()
ax.set_facecolor('#0B3D3B')

plt.plot(
    hourly_demand.index,
    hourly_demand.values,
    marker='o',
    linewidth=3,
    color='#39FFB6'
)

plt.fill_between(
    hourly_demand.index,
    hourly_demand.values,
    color='#39FFB6',
    alpha=0.15
)

plt.axvspan(
    14,
    17,
    color='#7CFFCB',
    alpha=0.15,
    label='Peak Demand Window'
)

plt.title(
    'Hourly Charging Demand Pattern',
    fontsize=20,
    color='white',
    pad=15
)

plt.xlabel(
    'Hour of Day',
    fontsize=12,
    color='white'
)

plt.ylabel(
    'Charging Sessions',
    fontsize=12,
    color='white'
)

plt.xticks(color='white')
plt.yticks(color='white')

plt.grid(
    alpha=0.2,
    color='white'
)

for spine in ax.spines.values():
    spine.set_color('white')

plt.legend(
    facecolor='#0B3D3B',
    edgecolor='white',
    labelcolor='white'
)

plt.tight_layout()

plt.savefig(
    'hourly_demand_presentation.png',
    dpi=300,
    bbox_inches='tight',
    facecolor='#0B3D3B'
)

plt.show()
```
