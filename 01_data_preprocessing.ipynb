```python
import pandas as pd
import numpy as np

df = pd.read_csv("caltech.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['connectionTime'] = pd.to_datetime(df['connectionTime'])

df['disconnectTime'] = pd.to_datetime(df['disconnectTime'])

df['doneChargingTime'] = pd.to_datetime(df['doneChargingTime'])
```

```python
df['connectionTime'].head()
```

```python
df['connectionTime'] = pd.to_datetime(
    df['connectionTime'],
    errors='coerce'
)

df['disconnectTime'] = pd.to_datetime(
    df['disconnectTime'],
    errors='coerce'
)

df['doneChargingTime'] = pd.to_datetime(
    df['doneChargingTime'],
    errors='coerce'
)
```

```python
df['connectionTime'] = pd.to_datetime(
    df['connectionTime'],
    utc=True
)

df['disconnectTime'] = pd.to_datetime(
    df['disconnectTime'],
    utc=True
)

df['doneChargingTime'] = pd.to_datetime(
    df['doneChargingTime'],
    utc=True
)
```

```python
df[['connectionTime','disconnectTime','doneChargingTime']].head()
```

```python
df.info()
```

```python
df['session_duration_hours'] = (
    df['disconnectTime'] - df['connectionTime']
).dt.total_seconds() / 3600

df['charging_duration_hours'] = (
    df['doneChargingTime'] - df['connectionTime']
).dt.total_seconds() / 3600
```

```python
df[['session_duration_hours',
    'charging_duration_hours']].describe()
```

```python
negative_charging = df[df['charging_duration_hours'] < 0]

print(len(negative_charging))
```

```python
df = df[df['charging_duration_hours'] >= 0]
```

```python
df['utilization_rate'] = (
    df['charging_duration_hours'] /
    df['session_duration_hours']
)
```

```python
df['utilization_rate'].describe()
```

```python
print(df['connectionTime'].min())
print(df['connectionTime'].max())
```

```python
df.to_csv(
    "processed_caltech.csv",
    index=False
)
```
